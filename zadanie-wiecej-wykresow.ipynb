```python
import numpy as np
import pandas as pd
```

```python
prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),
]

df = pd.DataFrame(prices, columns=["month", "pricePLN"])
df = df.set_index("month")
```

```python
df["priceUSD"] = [x / 4 for x in df["pricePLN"]]
df
```

```python
import matplotlib.pyplot as plt
plt.plot(df.index, df["priceUSD"], color="red", ls="dashed")
plt.title("Price of goods (USD)")
plt.xlabel("month")
plt.grid()
```
